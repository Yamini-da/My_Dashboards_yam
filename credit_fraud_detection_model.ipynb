```python
# import dependencies

import pandas as pd
import numpy as no
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

 DATA PREPARATION

```python
df_credit_data = pd.read_csv(r'C:\Users\WIKKI\Desktop\creditcard.csv')
```

```python
# check for null values
df_credit_data.isnull().sum()
```

```python
# check data stat pts
df_credit_data.groupby('Class').mean()
```

```python
# check fraudulent & legit numbers
df_credit_data['Class'].value_counts()
```

```python
# Assuming 'legit' and 'fraud' are subsets of your DataFrame
legit = df_credit_data[df_credit_data['Class'] == 0]
fraud = df_credit_data[df_credit_data['Class'] == 1]
```

```python
print(f'n values in legit,{legit['Class'].value_counts()},fraud ,{fraud['Class'].value_counts()}')
```

```python
# create 2 buckets of same size
legit = legit.sample(n=len(fraud))
# legit.shape[0]
```

```python
# after taking equal samples in both categories combine it

credit_data = pd.concat([legit,fraud],axis=0)
```

```python
credit_data['Class'].value_counts()
```

 MODEL Train & Test Data

```python
# split prediction column as y others as x
X= credit_data.drop(columns='Class')
Y= credit_data['Class']
```

```python
# splitting train and test data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
X_test = X_test.to_numpy()
Y_test = Y_test.to_numpy()
```

```python
# assigning model
model = LogisticRegression()
```

```python
# model fitting X

model.fit(X_train, Y_train)
```

```python
predict_Xtrain = model.predict(X_train)
score_Xtrain= accuracy_score(predict_Xtrain,Y_train)
```

```python
print(f'Training score of Credit-Fraud analysis is {score_Xtrain}')
```

```python
predict_Xtest= model.predict(X_test)
score_Xtest= accuracy_score(predict_Xtest,Y_test)
```

```python
print(f'Testing score of Credit-Fraud analysis is {score_Xtest}')
```
